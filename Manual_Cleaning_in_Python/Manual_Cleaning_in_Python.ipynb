```python
#Removing unnecessary index column
DS = pd.read_csv("IPL_Dataset.zip", dtype=str)
DS.drop('Unnamed: 0', axis=1, inplace=True)
print(DS)
```

```python
#Checking and removing duplicate rows
DS = pd.read_csv("IPL_Dataset.zip", low_memory=False)
print(DS.head())
duplicates = DS.duplicated().sum()
print("Duplicate Rows:", duplicates)
DS = DS.drop_duplicates()
```

```python
#Replacing missing values in wicket information
DS ['wicket_kind'] = DS ['wicket_kind'].fillna('No Wicket')
print(DS)
```

```python
#Replacing missing player_out values
DS['player_out'] = DS['player_out'].fillna('Not Out')
print(DS)
```

```python
#Standardize team names
DS['batting_team'] = DS['batting_team'].str.strip()
DS['bowling_team'] = DS['bowling_team'].str.strip()
print(DS)
```

```python
#Convert numeric columns to proper datatype
numeric_cols = ['runs_batter', 'runs_extras', 'runs_total']
DS[numeric_cols] = DS[numeric_cols].apply(pd.to_numeric, errors='coerce')
print(DS)
```

```python
#Convert boolean column properly
DS['valid_ball'] = DS['valid_ball'].astype(bool)
print(DS)
```
